```python
import numpy as np
import pandas as pd

df = pd.DataFrame(pd.read_csv('data_01.csv'))


#Independent variable
X = df.iloc[:,:-1].values

#Dependent variable
Y = df.iloc[:,-1]
# df.head()
print(X)



```

```python
###----------- pipline
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder,StandardScaler

num_cols = ['Age','Salary']
cat_cols = ['Country']

# imputer = SimpleImputer()
# encoder = OneHotEncoder()

cat_pipline = Pipeline([
       ('imputer', SimpleImputer(strategy='most_frequent')),
       ('encoder',OneHotEncoder(drop='first'))
]) 

num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])


###------combine pipline

from sklearn.compose import ColumnTransformer
preprocessor = ColumnTransformer([
    ('num',num_pipeline,num_cols),
    ('cat',cat_pipline,cat_cols)
])


###-------train-test split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)

# from sklearn
X_train = preprocessor.fit_transform(X_train)
X_test = preprocessor.transform(X_test)


```

```python
data = {
    'age': [25, np.nan, 35, 45, 55, 65, 75, 85, 95, 105],
    'salary': [50000, 60000, np.nan, 80000, 90000, 100000, 110000, 120000, np.nan, 140000],
    'gender': ['Male', 'Female', np.nan, 'Female', 'Male', 'Female', 'Male', 'Male', 'Female', 'Male'],
    'purchased': ['No', 'Yes', 'No', 'No', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes']
}

df = pd.DataFrame(data)

df.columns

```

```python
####-------- X-Y

X = df.drop('purchased',axis=1)
Y = df['purchased']

num_cols = ['age','salary']
cat_cols = ['gender','purchased']


####----------pipline making

num_pipeline = Pipeline([
               ('imputer',SimpleImputer(strategy='mean')),
                ('scaler',StandardScaler())
                
])

cat_pipline = Pipeline([
               ('imputer',SimpleImputer(strategy='most_frequent')),
               ('encoder',OneHotEncoder())
])

######## comnined pippline

preprocessor = ColumnTransformer([
    ('num',num_pipeline,num_cols),
    ('cat',cat_pipline,cat_cols)
])

###----
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)

##--- fit and transform
X_train = preprocessor.fit_transform(X_train)
X_test = preprocessor.transform(X_test)
# X_test



```

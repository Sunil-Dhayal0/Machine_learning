```python
import pandas as pd
from sklearn.metrics import precision_score, accuracy_score,recall_score
from sklearn.model_selection import train_test_split


heart_df = pd.DataFrame(pd.read_csv('heart.csv'))
heart_df.head()


####----feature selection
X = heart_df.drop('target',axis=1)
Y = heart_df['target']

##-> we don't use feature scale because output doesn't depends on the distance 
##   magnitude of output but in regresion there is matter

####--------train test split

X_train,X_test,Y_train,Y_test = train_test_split(
X,Y,test_size=0.2,random_state=42)




####--------model training -> noraml distribution - navie bayes
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train,Y_train)

# model prdict
y_p = model.predict(X_test)
print(y_p[1])
Y_test.iloc[1]

###--- model evaluation
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score,precision_score,recall_score,f1_score

cm = confusion_matrix(Y_test,y_p)
cm

print('accuracy_score: ',accuracy_score(Y_test,y_p))
print('precision_score: ',precision_score(Y_test,y_p))
# type-1 error false-positive   # not  spam email->  spam

print('recall_score :',recall_score(Y_test,y_p))
#type-2 error true-negative    # disease -> not disease



## hear recall is matter than  false-positive 
```

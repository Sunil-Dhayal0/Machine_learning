```python
import numpy as np
import pandas as pd

np.random.seed(23)

mu_vect1 = np.array([0,0,0])
cov_mat1 = np.array([[1,0,0],[0,1,0],[0,0,1]])
class1_samples = np.random.multivariate_normal(mu_vect1,cov_mat1,20)
df = pd.DataFrame(class1_samples,columns=['feature1','feature2','feature3'])
df['target'] = 0


mu_vect2 = np.array([0,0,0])
cov_mat2 = np.array([[1,0,0],[0,1,0],[0,0,1]])
class2_samples = np.random.multivariate_normal(mu_vect2,cov_mat2,20)
df1 = pd.DataFrame(class2_samples,columns=['feature1','feature2','feature3'])
df1['target'] = 1

df = pd.concat([df,df1],ignore_index=True)
df

###--------now convert this 3D into 2D

###-----step.1 mean center data
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()

df.iloc[:,0:3] = scalar.fit_transform(df.iloc[:,0:3])
# df


###----step.2 cov-matrix

cov_mat = df.iloc[:,0:3].cov()

###------step3. egine value and eign-vector
eigen_values,eigen_vectors = np.linalg.eig(cov_mat)

###----- selection of eigen_vectors
pc = eigen_vectors[0:2]  
# pc(2,3) pc.T = (3,2)
#   X(n,3)   pc.T(3,2)
transformed_df = np.dot(df.iloc[:,0:3],pc.T)

new_df = pd.DataFrame(transformed_df,columns=['pc1','pc2'])
new_df['target'] = df['target'].values
# new_df
pc
```
